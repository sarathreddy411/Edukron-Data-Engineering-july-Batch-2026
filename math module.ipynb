```python
# 30. Build a menu-driven calculator that uses arithmetic, comparison, logical, assignment, membership, identity, and bitwise operators wherever applicable.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("\n===== MENU =====")
print("1. Arithmetic")
print("2. Comparison")
print("3. Logical")
print("4. Assignment")
print("5. Membership")
print("6. Identity")
print("7. Bitwise")

choice = int(input("Enter your choice: "))

if choice == 1:
    print("Addition =", a + b)
    print("Subtraction =", a - b)
    print("Multiplication =", a * b)
    print("Division =", a / b)

elif choice == 2:
    print("a > b =", a > b)
    print("a < b =", a < b)
    print("a == b =", a == b)

elif choice == 3:
    print((a > 0) and (b > 0))
    print((a > 0) or (b > 0))
    print(not (a > b))

elif choice == 4:
    x = a
    x += b
    print("After += :", x)
    x -= b
    print("After -= :", x)

elif choice == 5:
    lst = [10, 20, 30, 40, 50]
    print(a in lst)
    print(a not in lst)

elif choice == 6:
    x = [1, 2, 3]
    y = x
    z = [1, 2, 3]
    print("x is y :", x is y)
    print("x is z :", x is z)
    print("x == z :", x == z)

elif choice == 7:
    print("AND =", a & b)
    print("OR =", a | b)
    print("XOR =", a ^ b)
    print("Left Shift =", a << 1)
    print("Right Shift =", a >> 1)

else:
    print("Invalid Choice")
```
