```python
# Create a list of 5 employee names and add a new employee using append().
emp=['sarath','santhosh','jaggu','lakshmi','narayana']
emp.append('manoj')
print(emp)

# Create a list of account numbers and add three new account numbers using extend().
acc=[101,102,103,104,105]
acc.extend([106,107,108])
print(acc)

# Insert a new customer at the beginning of a customer list using insert().
cust=['sarath','santhosh','jaggu','lakshmi','narayana']
cust.insert(0,"bharath")
print(cust)

# Insert a new customer at the end of a customer list using insert().
custt=['sarath','santhosh','jaggu','lakshmi','narayana']
custt.insert(-1,"bharath")
print(custt)

# Insert a transaction amount at index 3 in a list.
trac=[122,456,789,432,256]
trac.insert(3,999)
print(trac)

# Remove the last item from a list using pop().
l=[122,456,789,432,256]
l.pop()
print(l)

# Remove the customer "Rahul" from a customer list using remove().
cust=['sarath','santhosh','jaggu','Ramesh','lakshmi','narayana']
cust.remove('Ramesh')
print(cust)

# Count how many times the account type "Savings" appears in a list using count().
acc_type=['Savings','Current','Savings','Loan','Savings','Savings']
print(acc_type.count('Savings'))

# Find the index of "Current" in a list of account types using index().
acc_type=['Savings','Current','Savings','Loan','Savings','Savings']
print(acc_type.index('Current'))

# Reverse a list of transaction IDs using reverse().
trac=[122,456,789,432,256]
trac.reverse()
print(trac)

# Sort a list of account balances in ascending order using sort().
acc_bal=[12,45,67,4,32,56,9]
acc_bal.sort(reverse=False)
print(acc_bal)

# Sort a list of customer names in descending order using sort().
acc_bal=[12,45,67,4,32,56,9]
acc_bal.sort(reverse=True)
print(acc_bal)

# Make a copy of a list using copy() and verify that modifying the copied list does not affect the original list.
acc_bal=[12,45,67,4,32,56,9]
acc_bal_copy=acc_bal.copy()
print(acc_bal_copy)
acc_bal.append(100)
print(acc_bal)
print(acc_bal_copy)

# Clear all elements from a list using clear().
acc_bal=[12,45,67,4,32,56,9]
acc_bal.clear()
print(acc_bal)

# Append a nested list containing a new customer's details to an existing customer list.
cust=[1,2,3,4]
cust.append([[5,6,7,8],[23,4,5]])
print(cust)

# Extend a list of branch names with another list of new branches.
branches=['cse','ece','eee']
branches.extend(['mech','civil'])
print(branches)

# Remove the first occurrence of the balance 50000 from a list using remove().
acc_bal=[12,45,67,4,32,50000,9,50000]
acc_bal.remove(50000)
print(acc_bal)

# Pop the element at index 2 from a list and print the removed value.
acc_ball=[12,45,67,4,32,50000,9,50000]
b=acc_ball.pop(2)
# print(acc_ball)
print(b)

# Count how many times the city "Bangalore" appears in a list.
city=['chennai','bangalore','hyderabad','bangalore']
print(city.count('bangalore'))

# Reverse a list of bank names and print the result.
bank=['sbi','hdfc','icici']
bank.reverse()
print(bank)

# Create a list of loan amounts and sort it in descending order.
loans=[234,56,766,232,888]
loans.sort(reverse=True)
print(loans)

```
